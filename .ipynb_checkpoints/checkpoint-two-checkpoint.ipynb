## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
# Checking for possible outliers in the EU_Sales column.
plt.scatter(df['Year'],df['EU_Sales'])
plt.show()
```

```python
# checking for outliers in the NA_Sales column
plt.scatter(df['Year'],df['NA_Sales'])
plt.show()
```

```python
# Looking at overall global game sales over the course of the 40 years the dataset covers.
plt.bar(df['Year'],df['Global_Sales'])
plt.xlabel('Year')
plt.ylabel('Game Sales')
plt.title('Global Video Game Sales')
plt.show()
```
